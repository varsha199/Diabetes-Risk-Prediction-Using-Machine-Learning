```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
diabetes = pd.read_csv("data/raw/diabetes.csv")
print(diabetes.head())
```

## Numeric feature distributions

```python
diabetes.hist(bins=20, figsize=(12, 10), color='skyblue', edgecolor='black')
plt.suptitle("Feature Distributions", fontsize=16)
plt.show()
```

```python
sns.countplot(x="Outcome", data=diabetes, palette="Set2")
plt.title("Target Variable Distribution")
plt.show()
```

```python
corr = diabetes.corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
for col in diabetes.columns[:-1]:  # skip target column
    sns.boxplot(x=diabetes[col], color="lightblue")
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
sns.pairplot(diabetes, hue="Outcome")
plt.show()
```

```python
sns.boxplot(x="Outcome", y="Glucose", data=diabetes)
plt.title("Glucose levels by Outcome")
plt.show()
```
